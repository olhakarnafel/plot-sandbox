```python
import numpy as np
import matplotlib.pyplot as plt
import math

fig = plt.figure(figsize=(9, 4.5))

ax1 = plt.subplot(1, 2, 1)
ax2 = plt.subplot(1, 2, 2)

x = np.arange(0.0, 3.0, 0.01)
y = np.cos(2 * math.pi * x)

ax1.plot(x, y)
ax1.set_xlabel("period")
ax1.set_ylabel("cosine value")
ax1.set_title("cosine plot")

ax2.boxplot(y, vert=True)
ax2.set_xticks([])
ax2.set_ylabel("cosine value")
ax2.set_title("boxplot of cosine value")
plt.suptitle("cosine analysis")

plt.subplots_adjust(wspace = 0.5, top = 0.8)

plt.show()
```
